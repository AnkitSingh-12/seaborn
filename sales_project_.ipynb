# **Sales**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/sales_data.csv',encoding='ISO-8859-1')
df
```

**fetching null data from amount col**


```python
df[df['Amount'].isnull()]
df
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
df
```

```python
df.dropna(inplace=True)
df
```

```python
df.rename(columns={'Marital_Status':'Marital'},inplace=True)
df
```

```python
df['Marital'].replace({0:'Single',1:'Married'},inplace=True)
df
```

```python
df['Gender'].replace({'F':'Female','M':'Male'},inplace=True)
df
```

```python
plt.figure(figsize=(10,8),facecolor='yellow')
ax=sns.countplot(x='Gender', data=df, palette='rainbow', hue='Gender')
for i in ax.containers:
   ax.bar_label(i,)
ax=plt.gca()
ax.set_facecolor('red')
plt.title('Count of Males and Females', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(fontsize=10)
plt.show()
```

```python
df.groupby(['Gender','Age Group','State','Occupation','Product_Category','Marital']).sum()
```

```python

gender_spending = df.groupby('Gender')['Amount'].sum()
plt.figure(figsize=(8, 6),facecolor='blue')
ax =sns.barplot(x=gender_spending.index, y=gender_spending.values, palette='magma')
for i in ax.containers:
   ax.bar_label(i,)
ax=plt.gca()
ax.set_facecolor('z')
plt.title('Total Spending by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Spending')
plt.show()
```
